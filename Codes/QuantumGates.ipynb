```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
import matplotlib.pyplot as plt
```

BASIC LOGIC GATES[AND,NAND,XOR,OR]

```python
#nand gate
bell = QuantumCircuit(3,1)
bell.h(0)
bell.h(1)
bell.ccx(0,1,2)
bell.x(2)
bell.measure(2,0)
simulator = AerSimulator()
compiled_circuit = transpile(bell, simulator)
sim_result = simulator.run(compiled_circuit).result()
counts = sim_result.get_counts()
#bell.draw('mpl')
plot_histogram(counts)
```

```python
#OR GATE
qc = QuantumCircuit(3, 1)
qc.h(0) 
qc.h(1) 
qc.x(2) 
qc.cswap(0, 1,2)
qc.measure(1,0)
qc.draw('mpl')
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
sim_result = simulator.run(compiled_circuit).result()
counts = sim_result.get_counts()
#qc.draw('mpl')
plot_histogram(counts)
```

```python
#XOR GATE
qc = QuantumCircuit(2, 1)
qc.h(0) 
qc.h(1)  
qc.cx(0, 1)
qc.measure(1,0)
qc.draw('mpl')
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
sim_result = simulator.run(compiled_circuit).result()
counts = sim_result.get_counts()
#qc.draw('mpl')
plot_histogram(counts)
```

```python
#HALF ADDER 
q2 = QuantumCircuit(3,2)
q2.h(0)
q2.h(1)
q2.ccx(0,1,2)
q2.cx(0,1)
q2.measure(1,0)
q2.measure(2,1)
simulator = AerSimulator()
compiled_circuit = transpile(q2, simulator)
sim_result = simulator.run(compiled_circuit).result()
counts = sim_result.get_counts()
#plot_histogram(counts)
q2.draw("mpl")
```

```python
#FULL ADDER
qc = QuantumCircuit(5,5)
qc.h(0)
qc.h(1)
qc.h(2)
qc.measure(0,0)
qc.measure(1,1)
qc.measure(2,2)
qc.ccx(0,1,3)
qc.cx(0,1)
qc.ccx(1,2,3)
qc.cx(1,2)
qc.measure(3,4)
qc.measure(2,3)
qc.draw("mpl")
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
sim_result = simulator.run(compiled_circuit).result()
counts = sim_result.get_counts()
#plot_histogram(counts)
qc.draw('mpl')
```

```python
#Half subtractor
qc = QuantumCircuit(3,2)
qc.h(0)
qc.h(1)
qc.x(0)
qc.ccx(0,1,2)
qc.x(0)
qc.cx(0,1)
qc.measure(1,0)
qc.measure(2,1)
qc.draw('mpl')
simulator = AerSimulator()
compiled_circuit = transpile(qc, simulator)
sim_result = simulator.run(compiled_circuit).result()
counts = sim_result.get_counts()
#plot_histogram(counts)
qc.draw("mpl")
```
